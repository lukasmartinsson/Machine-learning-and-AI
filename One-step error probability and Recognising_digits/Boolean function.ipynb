```python
import numpy as np
import itertools
```

```python
#Initialize dimensions, trials, epochs, eta and a counter_list
nr_dimensions = [2,3,4,5]
nr_trials = 10**4
nr_epochs = 20
eta = 0.05
counter_list = []

#For every dimension listed
for j in range(len(nr_dimensions)):
    
    print(f'Dimension {j+2}')
    #Create all possible boolean inputs for each dimension
    bool_input = np.array(list(itertools.product([0, 1], repeat=nr_dimensions[j])))
    #Keeping track of which bool have been used
    used_bool = []
    #Tracker of all linearly separable boolean functions
    counter = 0

    #For every trial
    for i in range(nr_trials):
        if (i%1000) == 0:
            print('Trials', i)
        #Generate random boolean outputs
        bool_output = [np.random.choice([-1,1]) for _ in range(2**nr_dimensions[j])]

        if bool_output not in used_bool:
            #Initialize weight and bias
            w = np.random.normal(loc=0,scale=1, size=nr_dimensions[j])/np.sqrt(nr_dimensions[j])
            bias = 0

            #For every epoch
            for i in range(nr_epochs):
                #Initizlie the total error per epoch
                total_error = 0
                for mu in range(2**nr_dimensions[j]):
                    
                    #Calculate the output and error
                    output = np.sign(np.sum(np.multiply(w,bool_input[mu]))-bias)
                    error = bool_output[mu] - output

                    #Adjust the weight and bias
                    w += np.multiply(np.multiply(eta,error),bool_input[mu])
                    bias -= np.multiply(eta,error)

                    #Count the total error of the epoch
                    total_error += np.abs(np.sum(error))
    
                if total_error == 0:
                    counter += 1
                    break
                    
            used_bool.append(bool_output)

    counter_list.append(counter)

print(counter_list)


#Note here the output for 4 dimensions was 265 as opposed to 249 in the discussion, this is since I ran the code again before submitting it.
```
