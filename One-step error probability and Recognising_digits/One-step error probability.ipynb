```python
import random
import numpy as np

#Nr of patterns
patterns = [12,24,48,70,100,120]
#Nr of bits
N = 120

#Nr or trials
trials = 10**5

#Error counter list
error_counter_list = []

for i in range(len(patterns)):
    print(f'Pattern {i+1}')

    #Error counter
    error_counter = 0

    for k in range(int(trials/100)):
   
        if (k%20000) == 0:
            print('Trials', k)

        #Generate the random patterns
        random_patterns = np.array([[random.choice((1,-1)) for _ in range(N)] 
                                    for _ in range(patterns[i])]).T
        
        #Initialize weight w
        w = np.zeros((N, N))

        #Update the weights with Hebb's rule and set diagonal to zero
        for j in range(patterns[i]):
            w += np.outer(random_patterns[:,j], random_patterns[:,j]) / N
        np.fill_diagonal(w,0)
        
        #This uses the same random patterns 100 times, but since N=120 this will not    
        # matter for 10**5 trials but will speed up computation significantly
        for _ in range (100):
            #Randomly select a neuron
            rand_neuron = random.randint(0,N-1)

            #Randomly select a pattern
            rand_pattern = random.randint(0,patterns[i]-1)

            #Updates a neuron
            updated_neuron = np.sign(np.dot(w[rand_neuron],     
                            random_patterns[:,rand_pattern]))

            #Check if sgn = 0
            if updated_neuron == 0: updated_neuron = 1

            #Check if error
            if updated_neuron != random_patterns[rand_neuron][rand_pattern]: 
                error_counter += 1

    error_counter_list.append(error_counter/trials)

print(error_counter_list)
```
